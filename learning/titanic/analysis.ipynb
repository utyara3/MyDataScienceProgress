### Columns select

```python
ages = dataset['Age']
subset = dataset[['Name', 'Age']]
ages, subset
```

### Adding new columns

```python
dataset['Age_in_months'] = dataset['Age'] * 12
dataset['IsAdult'] = dataset['Age'] >= 18
dataset['AgeGroup'] = np.where(
    dataset['Age'] < 18, 'Child', np.where(
        dataset['Age'] < 60, 'Senior', 'Adult'
    )
)
dataset['NameLength'] = dataset['Name'].apply(len)
columns = ['Name', 'Age', 'Age_in_months', 'IsAdult', 'AgeGroup', 'NameLength']
dataset[columns].sample(3, random_state=RSTATE)
```

### Rename columns

```python
dataset = dataset.rename(columns={"Pclass": "PassangerClass", "SibSp": "SiblingsSpouses"})
dataset.head(3)
```

## Data Filtering

```python
women = dataset[dataset['Sex'] == 'female']
women.sample(3, random_state=RSTATE)
```

```python
old_men = dataset[(dataset['Sex'] == 'male') & (dataset['Age'] > 60)]
old_men
```

```python
women_or_older_than_18 = dataset[(dataset['Sex'] == 'female') | (dataset['Age'] > 18)]
women_or_older_than_18.sample(3, random_state=RSTATE)
```

```python
not_first_class = dataset[~(dataset['PassangerClass'] == 1)] # negation
not_first_class.sample(3, random_state=RSTATE)
```

```python
smith_family = dataset[dataset['Name'].str.contains('Smith')]
smith_family
```

```python
names_starting_with_a = dataset[dataset['Name'].str.lower().str.startswith('a')]
names_starting_with_a.sample(3, random_state=RSTATE)
```

```python
middle_aged = dataset[dataset['Age'].between(30, 50)]
middle_aged.sample(3, random_state=RSTATE)
```

```python
specific_ports = dataset[dataset['Embarked'].isin(["C", "Q"])]
specific_ports.sample(3, random_state=RSTATE)
```

## Sorting the data

```python
integer_ages = dataset[dataset['Age'].apply(
    lambda x: pd.notna(x) and x.is_integer()
)]
sorted_ages = integer_ages.sort_values('Age')
sorted_ages.tail(3)
```

```python
sorted_ages_in_descending_order = integer_ages.sort_values('Age', ascending=False)
sorted_ages_in_descending_order.head(3)
```

```python
sorted_ages_multiple = integer_ages.sort_values(['Age', 'PassangerClass'], ascending=[True, False])
sorted_ages_multiple.tail(3)
```

```python
sorted_index = integer_ages.sample(3, random_state=RSTATE).sort_index()
sorted_index
```

## Grouping Data

```python
age_by_class = dataset.groupby('PassangerClass')['Age'].mean()
age_by_class
```

```python
survival_by_sex = dataset.groupby('Sex')['Survived'].mean()
survival_by_sex
```

```python
#std - standart deviation ( )
stats_by_class = dataset.groupby('PassangerClass')['Age'].agg(['mean', 'median', 'std', 'count'])
stats_by_class
```

```python
detailed_stats = dataset.groupby(['PassangerClass', 'Sex']).agg(
    {
        'Age': ['mean', 'min', 'max'], 
        'Fare': 'mean', 
        'Survived': ['sum', 'mean']
    }
).round(2)
detailed_stats
```

## Handling Data Gaps

```python
dataset_dropped_na = dataset.dropna(subset=['Age'])
dataset_dropped_na
```

```python
dataset_filled = dataset.fillna(0)
dataset_filled_age = dataset['Age'].fillna(dataset['Age'].mean())
dataset_filled_previous = dataset.fillna(method='ffill') #fill in the previous value
```

## Changing the data structure

```python
pivot_survival = pd.pivot_table(
    dataset,
    values='Survived',
    index='PassangerClass',
    columns='Sex',
    aggfunc='mean'
)
pivot_survival
```

```python
melted = pd.melt(
    dataset,
    id_vars=['Name', 'PassangerClass'],
    value_vars=['Age', 'Fare'],
    var_name='Variable',
    value_name='Value'
)
print(melted.head())
print(melted.tail())
```

## A little bit of everything

```python
youngest = dataset.loc[dataset['Age'].idxmin()]
oldest = dataset.loc[dataset['Age'].idxmax()]
print(f"Youngest:\n{youngest}\n\nOldest:\n{oldest}")
```

```python
#     
dataset['Age_normalized'] = dataset.groupby('PassangerClass')['Age'].transform(
    lambda x: (x - x.mean()) / x.std()
)

print("   :")
print(dataset[['PassangerClass', 'Age', 'Age_normalized']].head())
```

```python
#     -   
dataset['Age_filled'] = dataset.groupby('PassangerClass')['Age'].transform(
    lambda x: x.fillna(x.median())
)

print("    :")
print(dataset[['PassangerClass', 'Age', 'Age_filled']].isnull().sum())
```

```python
#      ?
dataset['Fare_rank_in_class'] = dataset.groupby('PassangerClass')['Fare'].transform(
    lambda x: x.rank(ascending=False, method='dense')
)

print("     :")
print(dataset[['Name', 'PassangerClass', 'Fare', 'Fare_rank_in_class']].sort_values(['PassangerClass', 'Fare_rank_in_class']).head(10))
```

```python
#     vs 
fare_comparison = dataset.groupby('Survived')['Fare'].mean()
print(f"   : {fare_comparison[1]:.0f}$")
print(f"   : {fare_comparison[0]:.0f}$")
print("\n")
#   
class_survival = dataset.groupby('PassangerClass')['Survived'].mean()
print("   :")
for cls, rate in class_survival.items():
    print(f" {cls}: {rate:.1%}")
print("\n")
#    
dataset['FamilySize'] = dataset['SiblingsSpouses'] + dataset['Parch']
family_effect = dataset.groupby('FamilySize')['Survived'].mean()
print("    :")
print(family_effect.idxmax())
print("\n")
#   vs 
dataset['IsChild'] = dataset['IsAdult'] == False
child_survival = dataset.groupby('IsChild')['Survived'].mean()
print(f": : {child_survival[True]:.1%}, : {child_survival[False]:.1%}")
print("\n")
#   
embarked_survival = dataset.groupby('Embarked')['Survived'].mean()
print("     :")
for port, rate in embarked_survival.items():
    print(f"{port}: {rate:.1%}")
```
